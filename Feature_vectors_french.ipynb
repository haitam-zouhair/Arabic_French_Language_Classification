```
from gensim.models import LdaModel
from gensim import corpora
import pickle
import os
#root_path = "data"
#files_train = os.listdir(root_path)
corpus = pickle.load(open("/content/drive/My Drive/Colab Notebooks/workspace_henceforth_ensmr/workspace_henceforth/corpus.pkl", "rb"))
dictionary = corpora.Dictionary(corpus)
corps = [dictionary.doc2bow(text) for text in corpus]
lda_model = LdaModel.load('/content/drive/My Drive/Colab Notebooks/workspace_henceforth_ensmr/workspace_henceforth/Model saved/lda_model.model')
train_vecs = []
for i in range(len(corpus)):
    top_topics = lda_model.get_document_topics(corps[i], minimum_probability=0.0)
    topic_vec = [top_topics[i][1] for i in range(10)]
    #topic_vec.extend([corpus.iloc[i].real_counts]) 
    #topic_vec.extend([len(corpus.iloc[i].text)]) 
    train_vecs.append(topic_vec)
```
